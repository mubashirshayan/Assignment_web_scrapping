```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url ='https://countrycode.org/'
```

```python
response =requests.get(url)
response
```

```python
soup=BeautifulSoup(response.content,'html.parser')
```

```python
table = soup.find('table')
```

```python
dir(table)
```

```python
table_header =table.find_all('th')
```

```python
column_title =[i.text.strip() for i in table_header]
column_title
```

```python
df =pd.DataFrame(columns=column_title)
```

```python
df
```

```python
row_data =table.find_all('tr')
```

```python
for row in row_data[1:]:
    rows =row.find_all('td')
    allrowsdata=[j.text.strip() for j in rows]
    length=len(df)
    df.loc[length]=allrowsdata
```

```python
df
```

```python
df.to_csv("countries_data.csv",index=False)
```
